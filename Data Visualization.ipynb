# DMA_2024

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df1 = pd.read_excel('DMA_2024.xlsx', sheet_name='VESTAMID E40')
df2 = pd.read_excel('DMA_2024.xlsx', sheet_name='COVESTRO 83095')
df3 = pd.read_excel('DMA_2024.xlsx', sheet_name='TPU 2006')


x1 = df1['Temp.']
y1 = df1['tanδ']

x2 = df2['Temp.']
y2 = df2['tanδ']

x3 = df3['Temp.']
y3 = df3['tanδ']

plt.plot(x1,y1, label='VESTAMID E40', color='#991d85')
plt.plot(x2,y2, label='COVESTRO 83095')
plt.plot(x3,y3, label='TPU 2006')

plt.xlim(-20, 40)
plt.ylim(0, 0.2)

plt.title('tanδ Comparison between VESTAMID® E40 and TPU')
plt.xlabel('Temperature')
plt.ylabel('tanδ')
plt.legend()
plt.show()
```
