# DMA_2024

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df1 = pd.read_excel('https://github.com/zhongma95/Python/raw/main/DMA_2024.xlsx', sheet_name='VESTAMID E40')
df2 = pd.read_excel('https://github.com/zhongma95/Python/raw/main/DMA_2024.xlsx', sheet_name='COVESTRO 83095')
df3 = pd.read_excel('https://github.com/zhongma95/Python/raw/main/DMA_2024.xlsx', sheet_name='TPU 2006')


x1 = df1['Temp.']
y1 = df1['tanδ']

x2 = df2['Temp.']
y2 = df2['tanδ']

x3 = df3['Temp.']
y3 = df3['tanδ']

'''
plt.plot(x1,y1, label='VESTAMID E40', color='#991d85')
plt.plot(x2,y2, label='COVESTRO 83095')
plt.plot(x3,y3, label='TPU 2006')
'''
plt.plot(x1,y1, label='VESTAMID E40', color='#991d85', linewidth=3)
plt.plot(x2,y2, label='TPU 1')
plt.plot(x3,y3, label='TPU 2')


plt.xlim(-20, 40)
plt.ylim(0, 0.2)

plt.title('tanδ Comparison between VESTAMID® E40 and other TPU')
plt.xlabel('Temperature [℃]')
plt.ylabel('tanδ')
plt.legend()
plt.show()
```

# 3D Product Matrix

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

X = [40,55,62,68,71,73,75,73,79,80]
Y = [87,215,370,720,1100,1600,2270,2380,6800,12600]
Z = [90,196,350,539,900,1320,1890,1920,5900,12000]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X, Y, Z, c='r', marker='o')

ax.set_xlabel('Shore D')
ax.set_ylabel('Tensile Modulus [MPa]')
ax.set_zlabel('Flexural Modulus [MPa]')

plt.show()
```
